## Hyperparameter Tuning
Hyperparameters chosen for tuning are learning rate and batch size. These hyperparameters are quite important and tuning them can provide good results initially. Moreover, additional parameters were not chose in order to speed up training and avoid extra costs. hpo.py was created and is used to guide the tuning. Can be found in the project directory.

```python
hpo_ranges = {
    'batch_size' : IntegerParameter(32,64),
    'learning_rate' : ContinuousParameter(0.001,0.1)
}
```

# Creating sagemaker specific regex for Sagemaker to accept the values.

```python
metric_desc = [{
    'Name': "ValidationAccuracy",
    'Regex': 'ValidationAccuracy:([0-9\\.]+)'
}]
```

# Initializing the estimator using the GPU instance for timeliness.

```python
estimator = PyTorch(
    entry_point="hpo.py",
    role=role,
    instance_type="ml.g4dn.xlarge",
    instance_count=1,
    framework_version='1.8.0',
    py_version='py36'
)
```

```python
tuner = HyperparameterTuner(
    estimator=estimator,
    objective_metric_name='ValidationAccuracy',
    hyperparameter_ranges=hpo_ranges,
    metric_definitions=metric_desc,
    max_jobs=4,
    max_parallel_jobs=2
)
```

```python
tuner.fit({'training': inputs})
```

# Following hyperparamaters will yield best results according ti the tuner. The hyperparameters will be passed to the main training job.

```python
from sagemaker.analytics import HyperparameterTuningJobAnalytics

tuning_job = tuner.latest_tuning_job.name
analytics = HyperparameterTuningJobAnalytics(tuning_job)

best = analytics.dataframe().sort_values('FinalObjectiveValue', ascending=False).iloc[0]
print(best)
```

## Model Profiling and Debugging
Best hyperparameters after tuning were 46.0 for batch size and 0.00358 for the learning rate. These hyperparamters will passed dueing the estimator initialization. All the code including the added debugger and profiler can be found in train_model.py in the project directory.

# Importing additional libraries for debugging and prifoling. Creating rules to look for vanishing gradient, overfitting, overtraining and low quality weights. Intervals for debugger are set to 100 and for profiler to 500. Note: Profiler interval set to 500 can be less informative however, for this task it is chosen in order to fit the memory of the instance type. 

```python
from sagemaker.debugger import Rule, rule_configs
from sagemaker.debugger import DebuggerHookConfig, ProfilerRule, ProfilerConfig, FrameworkProfile

rules = [
    Rule.sagemaker(rule_configs.vanishing_gradient()),
    Rule.sagemaker(rule_configs.overfit()),
    Rule.sagemaker(rule_configs.overtraining()),
    Rule.sagemaker(rule_configs.poor_weight_initialization()),
]


debugger_hook_config = DebuggerHookConfig(
    hook_parameters={
        "train.save_interval": "100",
        "eval.save_interval": "10"
    }
)


profiler_config = ProfilerConfig(
    system_monitor_interval_millis=500,
    framework_profile_params=FrameworkProfile(num_steps=10)
)
```

# Initializing the estimator with best hyperparameters and rules for debugging and profiling

```python
estimator = PyTorch(
    entry_point="train_model.py",
    role=role,
    instance_type="ml.g4dn.xlarge",
    instance_count=1,
    framework_version="1.8.0",
    py_version="py36",
    hyperparameters={
        "batch_size": 46,
        "learning_rate": 0.00358,
        "epochs": 10
    },
    debugger_hook_config=debugger_hook_config,
    profiler_config=profiler_config,
    rules=rules
)
```

```python
estimator.fit({'training': inputs})
```

```python
from smdebug.trials import create_trial
import matplotlib.pyplot as plt

trial = create_trial(estimator.latest_job_debugger_artifacts_path())


print(trial.collections())


loss_tensors = trial.tensor_names(collection='losses')

if loss_tensors:

    loss_tensor_name = loss_tensors[0]
    

    tensor = trial.tensor(loss_tensor_name)
    steps = tensor.steps()
    values = [tensor.value(s) for s in steps]
    
    plt.figure(figsize=(10, 6))
    plt.plot(steps, values)
    plt.title(loss_tensor_name)
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()
else:
    print("no tensors")
```

# The loss curve from the debugger show no anomalous behaviour. The loos is decreasing rapidly at the start and decreases overall. Small flucatuations are normal as the model adjusts the learning rate. 

# In case there was an anomalous behaviour such as no decrease in loss at the start and overall. It could indicate that the learning rate was too high and the problem of canishing gradient additionaly it could indicate data input working incorrectly. In order to eliminate these type of problems overview the gradient collection on Sagemaker. Reduce the learning rate. Check the data pipeline to ensure correctness of the data. Inspect to check if certain images are corrupted.

# Outputting both CPU and GPU utilizations. Accesing the logs in S3 and parsing th JSONL files. Creating simple dataframe populated with mixed data.

```python
import pandas as pd
import json
from datetime import datetime
import matplotlib.pyplot as plt
import io

s3 = boto3.client('s3')

bucket = 'sagemaker-us-east-1-401657970710'
prefix = 'pytorch-training-2025-03-30-19-02-34-503/profiler-output/system/incremental/2025033019/'
response = s3.list_objects_v2(Bucket=bucket, Prefix=prefix)
json_data = [obj['Key'] for obj in response.get('Contents', []) if obj['Key'].endswith('.json')]

records = []

for key in json_data:
    response = s3.get_object(Bucket=bucket, Key=key)
    content = response['Body'].read().decode('utf-8')
    for line in content.strip().split('\n'):
        records.append(json.loads(line))

df = pd.DataFrame(records)
df['Datetime'] = pd.to_datetime(df['Timestamp'], unit='s')
display(df.head())
```

## Model Deploying

# Initializing the model from trained model in order to be deployed. New script had to be used to enable correct inference. Script can be found in the project directory as inference.py

```python
from sagemaker.pytorch import PyTorchModel
from sagemaker import get_execution_role
from sagemaker.serializers import IdentitySerializer
from sagemaker.deserializers import JSONDeserializer

role = get_execution_role()


model = PyTorchModel(
    model_data=estimator.model_data,  
    role=role,
    entry_point='inference.py',    
    framework_version='1.8.0',
    py_version='py36'
)
```

# Deploying the inference endpoint. Due to low number of predictions expected using the medium instance type. Serilizer is used to provide appropriate raw input of bytes to the model. Deserializer is used to output correct class.

```python
predictor = model.deploy(
    instance_type='ml.m5.large',
    initial_instance_count=1,
    serializer=IdentitySerializer(content_type='application/x-image'),
    deserializer=JSONDeserializer()
)
```

# One image of afghan hound breed was extracted from Google to get predictions. The image can be found in the project directory as afghan_hound.jpg. To get predictions endpoint is invoked to get deserialized output. Further the output is decoded and presented in the form of an index. Lastly the index is mapped to the appropriate class. Which in this case was correct.

```python
from torchvision import datasets
with open("afghan_hound.jpg", "rb") as f:
    image_bytes = f.read()

wrapped_image = io.BytesIO(image_bytes)

runtime = boto3.client("sagemaker-runtime")

response = runtime.invoke_endpoint(
    EndpointName="pytorch-inference-2025-03-31-03-32-57-980",  
    ContentType="application/x-image",
    Body=wrapped_image
)

result = response["Body"].read().decode()
print(result)
```
